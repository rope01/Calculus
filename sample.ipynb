## 1. Условие задачи

Рассматриваем задачу Коши
$$
\begin{cases}
y' = \dfrac{y}{x} - 2\sqrt{\dfrac{y}{x}}, \\
y(1) = 4,
\end{cases}
$$

Требуется:

1. Найти аналитическое решение $\varphi(x)$ данной задачи.
2. Численно решить задачу на отрезке $[x_0, b]$ следующими методами:
   1. методом Эйлера;
   2. методом Хойна;
   3. методом Рунге–Кутты 4-го порядка.
3. Выполнить интегрирование с постоянным шагом
   $$
   h = \frac{b - x_0}{n}, \quad x_0 = 1, \; b = 2, \; n = 10.
   $$
4. Построить на одной картинке графики $\varphi(x)$ и приближённых решений
   $\tilde\varphi^{(l)}(x)$ для всех трёх методов $l = 1,2,3$.
5. Найти абсолютные и относительные погрешности:
   $$
   \Delta_l = \max_{k \in \{1,\dots,n\}}
      \bigl|\varphi(x_k) - \tilde\varphi^{(l)}(x_k)\bigr|,
   $$
   $$
   \delta_l = \max_{k \in \{1,\dots,n\}}
      \frac{\bigl|\varphi(x_k) - \tilde\varphi^{(l)}(x_k)\bigr|}
           {\bigl|\tilde\varphi^{(l)}(x_k)\bigr|},
   $$
   где $x_k = x_0 + k h$.

В отчёт должны войти:

1. Постановка задачи.
2. Листинги программ.
3. Аналитическое решение задачи.
4. Графики решений.
5. Таблицы погрешностей.


## 2. Аналитическое решение

$$
\begin{cases}
y' = \dfrac{y}{x} - 2\sqrt{\dfrac{y}{x}}, \\
y(1) = 4.
\end{cases}
$$

замена
$$
w = \sqrt{\frac{y}{x}}.
\quad\Rightarrow\quad$$


$$\quad\Rightarrow\quad
\frac{y}{x} = w^2 \quad\Rightarrow\quad y = x w^2.
$$


т.к
$$
y = x w^2.
$$

$$
y' = w^2 + 2x w w'.
$$

из исходного
$$
y' = \frac{y}{x} - 2\sqrt{\frac{y}{x}}.
$$

падставляим $\frac{y}{x} = w^2$ и $\sqrt\frac{y}{x} = w$:
$$
y' = w^2 - 2 w.
$$

$$
w^2 + 2x w w' = w^2 - 2w.
$$

$$
2x w w' = -2w.
$$

так как $y(1) = 4$ и $\sqrt{y(1)/1} = 2$ $w \neq 0$ (это верно в окрестности точки $x = 1$)
$$
x w' = -1 \quad\Rightarrow\quad w' = -\frac{1}{x}.
$$

интегрируем:
$$
\int w'\,dx = \int -\frac{1}{x}\,dx
\quad\Rightarrow\quad
w = -\ln x + C,
$$

$y(1) = 4 \quad\Rightarrow\quad$
$$
w(1) = \sqrt{\frac{y(1)}{1}} = \sqrt{4} = 2.
$$
$$
w(1) = -\ln 1 + C = 0 + C = C \quad\Rightarrow\quad
$$

$$\quad\Rightarrow\quad
C = 2,
$$
и
$$
w = 2 - \ln x.
$$

### 2.4. Возвращаемся к функции $y$

связь между $y$ и $w$:
$$
y = x w^2.
$$

подставляем:
$$
y = x (2 - \ln x)^2.
$$

аналитическое решение:
$$
\varphi(x) = y = x (2 - \ln x)^2
$$

провеурка:
$$
\varphi(1) = 1 \cdot (2 - \ln 1)^2 = 1 \cdot 2^2 = 4,
$$
что совпадает с $y(1) = 4$.


## 3. Численные методы

Строим равномерную сетку на отрезке $[1,2]$:
$$
x_0 = 1, \quad x_k = x_0 + k h, \quad k = 0,1,\dots,n,
$$
где $h = \dfrac{b - x_0}{n} = 0.1$, $n = 10$.

### 3.1. Метод Эйлера

Метод Эйлера:
$$
\begin{cases}
x_{k+1} = x_k + h, \\
y_{k+1} = y_k + h\,f(x_k, y_k),
\end{cases}
\quad k = 0,\dots,n-1.
$$

Для нашего уравнения:
$$
y_{k+1} = y_k + h\left(
\frac{y_k}{x_k} - 2\sqrt{\frac{y_k}{x_k}}
\right).
$$

### 3.2. Метод Хойна

Метод Хойна:
$$
\begin{cases}
k_1 = f(x_k, y_k), \\
k_2 = f(x_k + h,\, y_k + h k_1), \\
y_{k+1} = y_k + \dfrac{h}{2}(k_1 + k_2), \\
x_{k+1} = x_k + h,
\end{cases}
\quad k = 0,\dots,n-1.
$$

Для нашего уравнения:
$$
k_1 = \frac{y_k}{x_k} - 2\sqrt{\frac{y_k}{x_k}},
$$
$$
k_2 = \frac{y_k + h k_1}{x_k + h}
      - 2\sqrt{\frac{y_k + h k_1}{x_k + h}}.
$$

### 3.3. Метод Рунге–Кутты 4-го порядка

Метод Рунге–Кутты 4-го порядка:
$$
\begin{cases}
k_1 = f(x_k, y_k), \\
k_2 = f\left(x_k + \dfrac{h}{2},\, y_k + \dfrac{h}{2}k_1\right), \\
k_3 = f\left(x_k + \dfrac{h}{2},\, y_k + \dfrac{h}{2}k_2\right), \\
k_4 = f\left(x_k + h,\, y_k + h k_3\right), \\
y_{k+1} = y_k + \dfrac{h}{6}\bigl(k_1 + 2k_2 + 2k_3 + k_4\bigr), \\
x_{k+1} = x_k + h,
\end{cases}
\quad k = 0,\dots,n-1.
$$

### 3.4. Погрешности

Точное решение:
$$
\varphi(x) = x(2 - \ln x)^2.
$$

Абсолютная погрешность в узле $x_k$:
$$
\varepsilon_k^{(l)} = \bigl|\varphi(x_k) - \tilde\varphi^{(l)}(x_k)\bigr|.
$$

Относительная погрешность:
$$
\delta_k^{(l)} =
\frac{\bigl|\varphi(x_k) - \tilde\varphi^{(l)}(x_k)\bigr|}
     {\bigl|\tilde\varphi^{(l)}(x_k)\bigr|}.
$$

Максимальные погрешности:
$$
\Delta_l = \max_{k=1,\dots,n} \varepsilon_k^{(l)}, \qquad
\delta_l = \max_{k=1,\dots,n} \delta_k^{(l)}.
$$


```python
import math
import matplotlib.pyplot as plt

x0 = 1.0
y0 = 4.0
b = 2.0
n = 10
h = (b - x0) / n

def f(x, y):
    return y / x - 2 * math.sqrt(y / x)

def phi(x):
    return x * (2 - math.log(x)) ** 2

def solve(method):
    xs = [x0]
    ys = [y0]
    x = x0
    y = y0
    for _ in range(n):
        if method == "euler":
            y = y + h * f(x, y)

        elif method == "heun":
            k1 = f(x, y)
            k2 = f(x + h, y + h * k1)
            y = y + h * (k1 + k2) / 2

        elif method == "rk4":
            k1 = f(x, y)
            k2 = f(x + h / 2, y + h * k1 / 2)
            k3 = f(x + h / 2, y + h * k2 / 2)
            k4 = f(x + h, y + h * k3)
            y = y + h * (k1 + 2*k2 + 2*k3 + k4) / 6

        x = x + h
        xs.append(x)
        ys.append(y)

    return xs, ys

xs, y_euler = solve("euler")
_,  y_heun  = solve("heun")
_,  y_rk4   = solve("rk4")

y_exact = [phi(x) for x in xs]

def max_errors(y_num):
    abs_err = []
    rel_err = []
    for x, ya, yn in zip(xs[1:], y_exact[1:], y_num[1:]):  # k = 1..n
        e = abs(ya - yn)
        abs_err.append(e)
        rel_err.append(e / abs(yn) if yn != 0 else float("inf"))
    return max(abs_err), max(rel_err)

Delta_euler, delta_euler = max_errors(y_euler)
Delta_heun,  delta_heun  = max_errors(y_heun)
Delta_rk4,   delta_rk4   = max_errors(y_rk4)

print("Максимальные погрешности (k = 1..n):")
print(f"Эйлер:       Δ = {Delta_euler:.6e}, δ = {delta_euler:.6e}")
print(f"Хойн:        Δ = {Delta_heun:.6e},  δ = {delta_heun:.6e}")
print(f"Рунге–Кутта: Δ = {Delta_rk4:.6e},  δ = {delta_rk4:.6e}")
print()

print("| k |  x_k  |  phi(x_k)  |  y_Euler  | |err_E|   |  y_Heun   | |err_H|   |  y_RK4    | |err_RK4|  |")
print("|---|-------|------------|-----------|----------|-----------|----------|-----------|----------|")
for k, (x, ye, yh, yr, ya) in enumerate(zip(xs, y_euler, y_heun, y_rk4, y_exact)):
    errE = abs(ya - ye)
    errH = abs(ya - yh)
    errR = abs(ya - yr)
    print(f"| {k:2d} | {x:5.2f} | {ya:10.6f} | {ye:9.6f} | {errE:8.6f} | "
          f"{yh:9.6f} | {errH:8.6f} | {yr:9.6f} | {errR:8.6f} |")

plt.plot(xs, y_exact, label="Точное решение")
plt.plot(xs, y_euler, "o--", label="Эйлер")
plt.plot(xs, y_heun, "s--", label="Хойн")
plt.plot(xs, y_rk4, "d--", label="Рунге–Кутта 4")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Сравнение точного и численных решений (вариант 14)")
plt.grid(True)
plt.legend()
plt.show()
```
