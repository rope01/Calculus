## Часть 1. Аналитический метод

### $$  x_n = \left( \frac{2n - 1}{n} \right) \arccos((-1)^n) $$

#### 1. Исследование сходимости

1. **Сходящиеся подпоследовательности:**

Рассмотрим поведение функции $$ \arccos((-1)^n) $$ для последовательности $x_n$. Значение $(-1)^n $ чередуется в зависимости от четности или нечетности $ n $. Если $ n $ четное, то $(-1)^n = 1$, а если нечетное, то $(-1)^n = -1$.
- **Для четных n:**

В этом случае $ (-1)^n = 1 $, и, следовательно:
$$
\arccos(1) = 0
$$
Это означает, что для всех четных значений $ n $ значение $\arccos((-1)^n)$ равно $0$. Мы обозначим эти значения как $ x_{2n} $, то есть:
$$
x_{2n} = \frac{2(2n) - 1}{2n} \cdot 0 = 0
$$
Таким образом, мы получили, что все элементы подпоследовательности $ \{x_{2n}\} $ равны нулю.

- **Для нечетных n:**

В этом случае $ (-1)^n = -1 $, и, следовательно:
$$
\arccos(-1) = \pi
$$
Это значит, что для всех нечетных значений $ n $ значение $\arccos((-1)^n)$ равно $\pi$. Мы обозначим эти значения как $ x_{2n-1} $, то есть:
$$
x_{2n-1} = \frac{2(2n - 1) - 1}{2n - 1} \cdot \pi
$$
Преобразуем это выражение:
$$
x_{2n-1} = \frac{4n - 3}{2n - 1} \cdot \pi
$$
Эта подпоследовательность $ x_{2n-1} $ стремится к $ 2\pi $ при $ n \to \infty $, так как $\frac{4n - 3}{2n - 1} \to 2$ при больших значениях $ n $.

**При $ n \to \infty $:**

Теперь у нас есть два предельных случая:
- $ x_{2n} \to 0 $ для четных значений $ n $, так как каждый элемент подпоследовательности $ x_{2n} = 0 $.
- $ x_{2n-1} \to 2\pi $ для нечетных значений $ n $, так как элементы подпоследовательности $ x_{2n-1} $ приближаются к $ 2\pi $.

Следовательно, выделяются две сходящиеся подпоследовательности:
$$
\{x_{2n}\} \to 0 \quad \text{и} \quad \{x_{2n-1}\} \to 2\pi
$$
Что бы узнать монотонна функция или нет, посчитаем первые три точки:$$\ $$

при n = 1 $\frac{2n - 1}{n} \arccos{(-1)^n} = \pi$ $$\ $$

при n = 2 $\frac{2n - 1}{n} \arccos{(-1)^n} = 0$ $$\ $$

при n = 3 $\frac{2n - 1}{n} \arccos{(-1)^n} = 2.5\pi$ $$\ $$

2. **Множество частичных пределов:**

Множество частичных пределов – это множество всех предельных значений, к которым сходятся подпоследовательности последовательности $ x_n $. В данном случае:
$$
\{0, 2\pi\}
$$
Так как одна подпоследовательность ($ \{x_{2n}\} $) сходится к $ 0 $, а другая подпоследовательность ($ \{x_{2n-1}\} $) сходится к $ 2\pi $, то частичные пределы последовательности $ x_n $ равны $ 0 $ и $ 2\pi $.

3. **Верхний и нижний пределы:**

- **Нижний предел (inf):** Наименьшее значение, к которому стремится последовательность $ x_n $, то есть:
$$
\lim \inf x_n = 0
$$
- **Верхний предел (sup):** Наибольшее значение, к которому стремится последовательность $ x_n $, то есть:
$$
\lim \sup x_n = 2\pi
$$

4. **Сходимость последовательности:**

Поскольку последовательность $ x_n $ имеет два различных предела (0 и $ 2\pi $), она **не сходится**. Для сходимости последовательности необходимо, чтобы все подпоследовательности сходились к одному и тому же пределу.

#### 2. Точные грани и дополнительные исследования

1. **Нахождение sup и inf:**

Точные грани последовательности $ x_n $ — это её наибольшие и наименьшие значения:
- Верхняя грань $ \sup x_n = 2\pi $, так как подпоследовательность $ x_{2n-1} $ приближается к $ 2\pi $.
- Нижняя грань $ \inf x_n = 0 $, так как подпоследовательность $ x_{2n} $ равна $ 0 $ для всех четных $ n $.

2. **Исследование монотонности:**

- Для четных $ n $ последовательность постоянна и равна 0, то есть:
$$
x_{2n} = 0
$$
- Для нечетных $ n $:
$$
x_{2n-1} = \left(2 - \frac{1}{2n - 1}\right) \cdot \pi
$$
Заметим, что $\frac{1}{2n - 1}$ убывает при увеличении $ n $. Это значит, что $ x_{2n-1} $ монотонно возрастает и приближается к $ 2\pi $ при больших значениях $ n $.

Таким образом, последовательность $ x_{2n-1} $ является монотонно возрастающей, а последовательность $ x_{2n} $ — постоянной.

#### 3. Наибольший и наименьший элементы

1. **Наименьший элемент:**
$$
m = \inf x_n = 0
$$
Этот элемент достигается для четных значений $ n $ (все элементы подпоследовательности $ x_{2n} $ равны $ 0 $).

2. **Наибольший элемент:**
$$
M = \sup x_n = 2\pi
$$
Значение $ 2\pi $ является верхней гранью последовательности, к которой приближается $ x_{2n-1} $ при нечетных значениях $ n $, однако это значение никогда не достигается.

Таким образом, последовательность $ x_n $ имеет наименьший элемент $ 0 $, но не имеет наибольшего элемента.

#### 4. Выбор сходящейся подпоследовательности

1. **Выбор:** Рассмотрим подпоследовательность $ x_{2n} $, которая состоит из элементов с четными индексами и для которой известно, что:
$$
x_{2n} = 0
$$
Эта подпоследовательность является сходящейся к 0.

2. **Определение предела:**
Для подпоследовательности $ x_{2n} $ предел равен:
$$
\lim_{n \to \infty} x_{2n} = 0
$$

3. **Доказательство предела:**

Чтобы доказать, что $ \lim_{n \to \infty} x_{2n} = 0 $, воспользуемся $\varepsilon$-определением предела. Нам нужно показать, что для любого $\varepsilon > 0$ существует такое число $ N $, что для всех $ n > N $ выполняется неравенство:
$$
|x_{2n} - 0| < \varepsilon
$$
Так как $ x_{2n} = 0 $ для всех четных $ n $, то:
$$
|x_{2n} - 0| = 0 < \varepsilon
$$
Таким образом, для любого $\varepsilon > 0$ и любого $ n $ выполняется это условие, и, следовательно, $ x_{2n} $ сходится к $ 0 $.


```python
import numpy as np  # Импортирую библиотечку numpy для работы с числовыми массивчиками
import matplotlib.pyplot as plt  # Импортирую библиотечку для построения графиковиков

# Определение последовательности
n = np.arange(1, 101)  # Создаю массивчик из первых 100 натуральных чиселок от 1 до 100
x_n = ((2 * n - 1) / n) * (np.arccos((-1) ** n))  # Вычисляю значенье последовательности x_n по заданной формулке

# Аналитические границы
sup_xn = 2 * np.pi   # Задаю супремум последовательности, равную 2π
inf_xn = np.min(x_n)   # Находю инфимум последовательности, это минимальное значение x_n
limsup_xn = sup_xn     # Верхний асимптотический предел тоже равен 2π
liminf_xn = np.min(x_n)  # Нижний асимптотический предел равен минимальному значению последовательности

# Определение четных n для подпоследовательности
n_even = n[n % 2 == 0]  # Выбираю четные чиселки из массива n
x_n_even = x_n[n % 2 == 0]  # Получаю соответствующие значенье последовательности для четных n

# Создание графиков
plt.figure(figsize=(12, 12))  # Увеличиваю размерчик графиков, чтобы они были лучше видны

# Первый график: Полная последовательность
plt.subplot(2, 1, 1)  # Создаю подграфик: 2 строки, 1 столбик, я работаю с первым графиком
plt.axhline(sup_xn, color='red', linestyle=':', label='Супремум $sup(x_n)$', alpha=0.7, linewidth=2)  # Рисую горизонтальную линию для супремума
plt.axhline(inf_xn, color='green', linestyle='--', label='Инфимум $inf(x_n)$', alpha=0.7, linewidth=2)  # Рисую горизонтальную линию для инфимума
plt.axhline(limsup_xn, color='purple', linestyle='-', label='Верхний предел $limsup(x_n)$', linewidth=2)  # Рисую линию для верхнего предела
plt.axhline(liminf_xn, color='orange', linestyle='--', label='Нижний предел $liminf(x_n)$', linewidth=2)  # Рисую линию для нижнего предела
plt.plot(n, x_n, 'o-', label=r'Последовательность $x_n = \frac{(2n-1)}{n} \cdot \arccos((-1)^n)$', color='blue')  # Строю график для полной последовательности
plt.scatter([n[np.argmax(x_n)]], [sup_xn], color='red', s=100, zorder=5, label='Точка $sup(x_n)$')  # Отмечаю точку супремума на графике
plt.scatter([n[np.argmin(x_n)]], [inf_xn], color='green', s=100, zorder=5, label='Точка $inf(x_n)$')  # Отмечаю точку инфимума на графике
plt.ylim(-1, 7)  # Устанавливаю пределы по оси y от -1 до 7
plt.title('График полной последовательности')  # Заголовок графика
plt.xlabel('n')  # Подпись оси x
plt.ylabel('$x_n$')  # Подпись оси y
plt.legend(loc='upper left')  # Добавляю легенду в верхний левый угол
plt.grid()  # Включаю сеточку на графике

# Отображение графиков
plt.tight_layout()  # Убираю лишние отступы для улучшения внешнего вида графиков
plt.show()  # Показываю графики
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Определение последовательности
def sequence(n):
    return ((2 * n - 1) / n) * np.arccos((-1) ** n)

# Функция для расчета эпсилон-окрестностей и поиска точек выхода за границы
def epsilon_info(eps: float) -> tuple:
    N = 100  # Количество точек для отображения
    n_values = np.arange(1, N + 1)
    seq_values = sequence(n_values)

    # Определяем точные верхние и нижние грани
    sup_xn = np.max(seq_values)
    inf_xn = np.min(seq_values)

    # Определяем границы эпсилон-окрестностей
    sup_eps = sup_xn - eps
    inf_eps = inf_xn + eps

    # Ищем первые индексы m, где последовательность выходит за границы эпсилон-окрестностей
    m_sup = np.where(seq_values > sup_eps)[0][0] + 1
    m_inf = np.where(seq_values < inf_eps)[0][0] + 1

    return sup_xn, inf_xn, sup_eps, inf_eps, m_sup, m_inf, n_values, seq_values

# Задаем значения epsilon для визуализации
epsilons = [0.01, 0.001, 0.0001]
colors = ['green', 'blue', 'red']

# Построение отдельных графиков для каждого значения epsilon
for i, epsilon in enumerate(epsilons):
    sup_xn, inf_xn, sup_eps, inf_eps, m_sup, m_inf, n_values, seq_values = epsilon_info(epsilon)

    # Создаем график для текущего значения epsilon
    plt.figure(figsize=(10, 5))
    plt.plot(n_values, seq_values, label="x_n", color="royalblue", marker='o', markersize=4, linewidth=1)

    # Линии эпсилон-окрестности
    plt.axhline(sup_eps, color='magenta', linestyle='--', label=f'Верхняя граница ε={epsilon}')
    plt.axhline(inf_eps, color='purple', linestyle='--', label=f'Нижняя граница ε={epsilon}')
    
    # Точки выхода за границы эпсилон-окрестностей
    plt.plot(m_inf, seq_values[m_inf - 1], 'go', label=f"m для ε={epsilon}: {m_inf}", markersize=10)
    plt.plot(m_sup, seq_values[m_sup - 1], 'bo', label=f"m для ε={epsilon}: {m_sup}", markersize=10)

    # Настройки графика
    plt.title(f"Эпсилон-окрестности и точки последовательности для ε = {epsilon}")
    plt.xlabel("Номер n")
    plt.ylabel("Значение x_n")
    plt.legend()
    plt.grid(True)

    # Отображение графика
    plt.show()
```
