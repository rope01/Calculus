# Производные функции $$(\cos(\pi/4 + x))^2$$
## Математическое выражение функции
<hr/>
Рассмотрим функцию:
$$
f(x) = (cos(pi/4 + x))^2
$$

### 1. Первая производная:
Используем формулу произведения:
$$
f'(x) = -2 \sin\left(x + \frac{\pi}{4}\right) \cos\left(x + \frac{\pi}{4}\right) = -\sin\left(2x + \frac{\pi}{2}\right).
$$
Или:
$$
f'(x) = -\cos(2x).
$$

### 2. Вторая производная:
Раскроем выражение:
$$
f''(x) = 2 \left( \sin^2\left(x + \frac{\pi}{4}\right) - \cos^2\left(x + \frac{\pi}{4}\right) \right) = -2 \cos\left(2x + \frac{\pi}{2}\right).
$$
Или:
$$
f''(x) = -2 \sin(2x).
$$

### 3. Третья производная:
Упростим формулу через произведение:
$$
f'''(x) = 4 \sin\left(x + \frac{\pi}{4}\right) \cos\left(x + \frac{\pi}{4}\right) = 4 \sin\left(2x + \frac{\pi}{2}\right).
$$
Или:
$$
f'''(x) = 4 \cos(2x).
$$

### 4. Четвёртая производная:
Используем стандартное тождество:
$$
f''''(x) = 4 \left( \cos^2\left(x + \frac{\pi}{4}\right) - \sin^2\left(x + \frac{\pi}{4}\right) \right) = 8 \cos\left(2x + \frac{\pi}{2}\right).
$$
Или:
$$
f''''(x) = -8 \sin(2x).
$$

# Доказательство индукцией для производных функции $ f(x) = (\cos(\pi/4 + x))^2 $

## Формулировка гипотезы

Предположим, что для произвольного $ n $ (при $ n \geq 1 $) существует такая формула для $ n $-й производной:

$$
f^{(n)}(x) = (-1)^n \cdot 2^n \sin\left(x + \frac{\pi}{4}\right) \cos\left(x + \frac{\pi}{4}\right)
$$

## База индукции (для $ n = 1 $)

Для первой производной имеем:

$$
f'(x) = -2 \sin\left(x + \frac{\pi}{4}\right) \cos\left(x + \frac{\pi}{4}\right)
$$

Это совпадает с нашей гипотезой для $ n = 1 $, так как:

$$
f^{(1)}(x) = (-1)^1 \cdot 2^1 \sin\left(x + \frac{\pi}{4}\right) \cos\left(x + \frac{\pi}{4}\right)
$$

Следовательно, база индукции выполнена.

## Шаг индукции

Предположим, что гипотеза верна для $ n = k $, то есть:

$$
f^{(k)}(x) = (-1)^k \cdot 2^k \sin\left(x + \frac{\pi}{4}\right) \cos\left(x + \frac{\pi}{4}\right)
$$

Нужно доказать, что гипотеза верна для $ n = k + 1 $.

Для этого найдем $ (k+1) $-ю производную:

$$
f^{(k+1)}(x) = \frac{d}{dx} \left( (-1)^k \cdot 2^k \sin\left(x + \frac{\pi}{4}\right) \cos\left(x + \frac{\pi}{4}\right) \right)
$$

Используем правило произведения для производных:

$$
f^{(k+1)}(x) = (-1)^k \cdot 2^k \left[ \cos^2\left(x + \frac{\pi}{4}\right) - \sin^2\left(x + \frac{\pi}{4}\right) \right]
$$

Это можно записать как:

$$
f^{(k+1)}(x) = (-1)^{k+1} \cdot 2^{k+1} \sin\left(x + \frac{\pi}{4}\right) \cos\left(x + \frac{\pi}{4}\right)
$$

Таким образом, мы доказали, что гипотеза верна для $ n = k + 1 $.

## Заключение

Поскольку база индукции верна, и мы показали, что если гипотеза верна для $ n = k $, то она верна и для $ n = k + 1 $, по принципу математической индукции, мы доказали, что:

$$
f^{(n)}(x) = (-1)^n \cdot 2^n \sin\left(x + \frac{\pi}{4}\right) \cos\left(x + \frac{\pi}{4}\right)
$$

для всех $ n \geq 1 $.

Разложим функцию $ f(x) = (\cos(\pi/4 + x))^2 $ в ряд Тейлора порядка $ n $ вокруг точки $ x_0 = 0 $. Формула ряда Тейлора имеет вид:

$$
f(x) = \sum_{k=0}^{n} \frac{f^{(k)}(x_0)}{k!} (x - x_0)^k.
$$

Где $ f^{(k)}(x_0) $ — значение $ k $-й производной функции $ f(x) $ в точке $ x_0 $.

## Часть 2. Построение ряда Тейлора аналитически

### Ряд Тейлора до порядка $n$:
Ряд Тейлора вычисляется как:
$$
f(x) = \sum_{k=0}^n \frac{f^{(k)}(0)}{k!} x^k
$$

Найденные значения производных:
- $f(0) = \frac{1}{2}$,
- $f'(0) = -1$,
- $f''(0) = 0$,
- $f'''(0) = 4$,
- $f''''(0) = 0$.

Подставляем значения производных в формулу ряда Тейлора:
$$
f(x) = \frac{f(0)}{0!} + \frac{f'(0)}{1!}x + \frac{f''(0)}{2!}x^2 + \frac{f'''(0)}{3!}x^3 + \dots
$$

Подставляем:
$$
f(x) = \frac{1}{2} - x + \frac{2}{6}x^3 + \dots
$$

### Итоговый многочлен Тейлора до третьего порядка:
$$
f(x) = \frac{1}{2} - x + \frac{2}{3}x^3.
$$

Многочлен Тейлора:
$$
f(x) = \frac{1}{2} - x + \frac{2}{3}x^3 + \dots
$$

## Часть 3. Построение ряда Тейлора через преобразования

### Раскладка через известные многочлены Тейлора

$$
\left( \cos \left( \frac{\pi}{4} + x \right) \right)^2
$$

Используем тождество для квадрата косинуса:

$$
\cos^2(\theta) = \frac{1 + \cos(2\theta)}{2}
$$

Применим это к выражению:

$$
\left( \cos \left( \frac{\pi}{4} + x \right) \right)^2 = \frac{1 + \cos \left( 2 \left( \frac{\pi}{4} + x \right) \right)}{2}
$$

Упростим аргумент косинуса:

$$
= \frac{1 + \cos \left( \frac{\pi}{2} + 2x \right)}{2}
$$

Таким образом, получаем:

$$
= \frac{1}{2} + \frac{1}{2} \cos \left( \frac{\pi}{2} + 2x \right)
$$

Ряд для $\cos(x)$:
$$
\cos(x) = 1 - \frac{x^2}{2!} + \frac{x^4}{4!} - \dots
$$

Подставим $x = 2x + \frac{\pi}{2}$:
$$
f(x) = \frac{1}{2} - x + \frac{2}{3}x^3 + \dots
$$

## Часть 4
### Остаточный член формулы Тейлора

Формула остаточного члена Тейлора в форме Лагранжа:

$$
R_n(x, x_0) = \frac{f^{(n+1)}(\xi)}{(n+1)!}(x - x_0)^{n+1},
$$

где $ \xi \in (x_0, x) $.

Подставим $ x_0 = 0 $, $ x = a = -0.1 $:

$$
R_n(a) = \frac{f^{(n+1)}(\xi)}{(n+1)!}a^{n+1}.
$$

---

### Максимальная оценка $ f^{(n+1)}(\xi) $

На отрезке $\xi \in (0, a) $, тригонометрические функции $|\sin и \cos $ достигают максимального значения 1.  
Следовательно:

$$
|f^{(n+1)}(\xi)| \leq |C_n|,
$$

где $ C_n $ — коэффициент перед тригонометрической функцией.

Например:
- Для $n = 3 $: $ C_3 = 4 $,
- Для $ n = 5 $: $ C_5 = 8 $.

---

### Остаточный член для $ a = -0.1 $

Подставляем значения в формулу:

$$
R_n(a) = \frac{|C_n| \cdot |a|^{n+1}}{(n+1)!}.
$$

1. Для $n = 3$:

$$
R_3(a) = \frac{4 \cdot (0.1)^4}{4!} = \frac{4 \cdot 0.0001}{24} = \frac{0.0004}{24} \approx 1.67 \cdot 10^{-5}.
$$

2. Для $ n = 5 $:

$$
R_5(a) = \frac{8 \cdot (0.1)^6}{6!} = \frac{8 \cdot 0.000001}{720} = \frac{0.000008}{720} \approx 1.11 \cdot 10^{-8}.
$$

---

### Итог

- Для $ \Delta_1 = 10^{-3} $, минимальное $ n_1 = 3 $.
- Для $\Delta_2 = 10^{-6} $, минимальное $ n_2 = 5 $.


```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline

# Основная функция
def function(x):
    return ((np.cos((np.pi / 4) + x)))**2

# Производные от функции
def get_n_derivative(n, x):
    if n == 0:
        return function(x)
    elif n == 1:
        return -2 * np.cos(np.pi / 4 + x) * np.sin(np.pi / 4 + x)
    elif n == 2:
        return -2 * (np.cos(2 * x + np.pi / 2))
    elif n == 3:
        return 4 * np.sin(2 * x + np.pi / 2)
    elif n == 4:
        return 8 * np.cos(2 * x + np.pi / 2)
    elif n == 5:
        return -16 * np.sin(2 * x + np.pi / 2)
    else:
        raise ValueError()

# Многочлен Тейлора
def taylor_polynom(x, x0, n):
    result_sum = 0
    for k in range(n + 1):
        derivative_at_x0 = get_n_derivative(k, x0)
        result_sum += (derivative_at_x0 / math.factorial(k)) * ((x - x0) ** k)
    return result_sum

X = np.linspace(-3 * np.pi, 3 * np.pi, 1000) 
x0 = -0.1  # Точка разложения
# построение графика
plt.figure(figsize=(14, 8))
plt.plot(X, function(X), label="Main Function", color="black", lw=2)
plt.title("Main Function", fontsize=22)
plt.xlabel("$x$", fontsize=16)
plt.ylabel("$f(x)$", fontsize=16)
plt.axhline(0, color="gray", linestyle="--", lw=1)
plt.grid(True)
plt.ylim(0, 1.1)
plt.legend()
plt.show()

# Графики основной функции с тейлором
for n in range(1, 6):  # От 1-го до 5-го порядка
    plt.figure(figsize=(14, 8))
    # Основная функция
    plt.plot(X, function(X), label="Main Function", color="black", lw=2)
    # Многочлен Тейлора
    plt.plot(X, taylor_polynom(X, x0, n), label=f"Taylor (n={n})", linestyle="--", lw=2, color="blue")
    plt.title(f"Main Function and Taylor Polynomial (n={n})", fontsize=22)
    plt.xlabel("$x$", fontsize=16)
    plt.ylabel("$f(x)$ / $P_n(x)$", fontsize=16)
    plt.axhline(0, color="gray", linestyle="--", lw=1)
    plt.grid(True)
    plt.ylim(-0.5, 1.5)
    plt.legend()
    plt.show()

```

```python
# Точка разложения
a = -0.1

# Вычисление точного значения функции в точке a
exact_value = function(a)
print(f"f(a) (точное значение): {exact_value:.16f}")

# Вычисление значений многочлена Тейлора 3-го порядка в точке a
taylor_3rd = taylor_polynom(a, x0, 3)
print(f"P3(a) (Тейлор, 3-й порядок): {taylor_3rd:.16f}")

# Вычисление погрешности для 3-го порядка
R_n1 = abs(exact_value - taylor_3rd)
print(f"R_n1 (погрешность 3-го порядка): {R_n1:.16f}")

# Проверка, меньше ли погрешность заданного порога 10^-3
print(f"R_n1 < 10**(-3): {R_n1 < 10**(-3)}")

# Вычисление значений многочлена Тейлора 5-го порядка в точке a
taylor_5th = taylor_polynom(a, x0, 5)
print(f"P5(a) (Тейлор, 5-й порядок): {taylor_5th:.16f}")

# Вычисление погрешности для 5-го порядка
R_n2 = abs(exact_value - taylor_5th)
print(f"R_n2 (погрешность 5-го порядка): {R_n2:.16f}")

# Проверка, меньше ли погрешность заданного порога 10^-6
print(f"R_n2 < 10**(-6): {R_n2 < 10**(-6)}")
```
